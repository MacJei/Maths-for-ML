### Задача 1
Рассмотрим уравнение
$$2x_1 + 3x_2 - 5x_3 = 0.$$

Оно задаёт некоторую плоскость. Найти координаты двух линейно независимых векторов, лежащих в этой плоскости.

**Подсказка.** Искомые векторы можно найти в виде $(x_1^1, 1, 0)$ и $(x_1^2, 0, 1)$ (верхний индекс — это номер вектора, а не степень). Докажите, что любые такие векторы линейно независимы (какими бы ни были $x_1^1$ и $x_1^2$). Найдите $x_1^1$ и $x_1^2$ из уравнения.

**Решение**    
Пусть искомые векторы $v_1$ и $v_2$ имеют координаты $(x_1^1, 1, 0)$ и $(x_1^1, 0, 1)$ соответсвенно. Тогда:   
для $v_1$ можно записать:    
$2x_1^1 + 3 = 0$, отсюда $x_1 = \frac{-3}{2}$    
для  $v_2$ можно записать:    
$2x_1^2 + 5 = 0$, отсюда $x_1 = \frac{-5}{2}$     
Докажем, что любые векторы $v_1$ и $v_2$ заданного вида линейно независимы. Иначе можно сказать, что компоненты векторов не пропорциональны. Учитывая, что один из компонент вектора $v_1$ равен нулю,  то для вектора $v_2$ соответсвующий компонент должен  быть тоже равен нулю в случае их зависимости, так как это не выполняется $1 \neq 0$, то векторы линейно независимы.

*Линейным подпространством* $V \subset \mathbb R^n$ называется такое подмножество векторов, что для любых $u, v\in V$ и любого скаляра $\lambda \in \mathbb R$:

- $u + v \in V$;
- $\lambda u \in V$.

*Аффинным подпространством* $W \subset \mathbb R^n$  множество векторов $\{u_0 + v \mid v \in  V\}$, получаемых из некоторого фиксированного вектора $u_0$ прибавлением всевозможных векторов из линейного подпространства $V$. Иными словами, аффинное подпространство — это результат сдвига линейного подпространства на некоторый фиксированный вектор.

### Задача 5 (*)
Доказать, что для любой прямоугольной матрицы $A$ множество всех решений системы $Ax=0$ является линейным подпространством (если оно непусто, то есть если хоть какие-то решения у системы есть), а множество всех решений системы $Ax=b$ является аффинным подпространством. Как связаны решения этих двух систем?

**Решение**    
Положим, что система линеный уравнений вида $Ax = 0$ имеет решение $x = (x_1, x_2, ..., x_n) \in V$, $V \in \mathbb R^n$, иначе говоря, решение системы образует подпространсвто решений $V \in \mathbb R^n$, так что $x = (x_1, x_2, ..., x_n) \in \mathbb V$ такое, что $Ax = 0$, положим, теперь, что накоторые векторы $u, v \in V$.     
$v = (x_1^2, x_2^v,\dots, x_n^v)$, $u = (x_1^u, x_2^u,\dots, x_n^u)$   
$Av = 0$ и $Au = 0$    
Проверим, является ли подпространство $V$ линейным:    
1. $u + v = (x_1^u + x_1^v,\ x_2^u + x_2^v,\ \dots,\ x_n^u + x_n^v)$    
$A(u + v) = Au + Av = 0$
2. $\lambda u = (\lambda x_1^u , \lambda x_2^u , \dots, \lambda x_n^u)$   
$A(\lambda u) = A\lambda u = \lambda (Av) = 0$    

Аналогичные рассуждения проведем, для уравнения вида $Ax = b$, пусть теперь пространство решений системы $V_b \in \mathbb R^n$ положим, теперь, что накоторые векторы $u_b, v_b \in V_b$    
$Av_b = b$ и $Au_b = b$    
Проверим, является ли подпространство $V_b$ линейным:    
1. $u_b + v_b = (x_1^{u_b} + x_1^{v_b},\ x_2^{u_b} + x_2^{v_b},\ \dots,\ x_n^{u_b} + x_n^{v_b})$    
$A(u_b + v_b) = Au_b + Av_b = 2b \neq b$
2. $\lambda u_b = (\lambda x_1^{u_b} , \lambda x_2^{u_b} , \dots, \lambda x_n^{u_b})$   
$A(\lambda u_b) = A\lambda u_b = \lambda (Av_b) = \lambda b \neq b$   

Докажем, что пространство $V_b$ аффинное. Пусть $Ax = 0$ задает подпространство $V \in \mathbb R^n$, как было показано выше, оно являтся линейным, и пусть задано подпространство $V_b \in \mathbb R^n$ как решение системы $Ax = b$. Пусть существует некторое решение $x_0 \in V_b$, тогда $Ax_0 = b$, прибавление нуля не изменит это выражение, тогда рассмотрим систему $Ax_0 + Ax = b$ иначе можно записать $A(x_0 + x) = b$. Решения данной системы являеются вектора вида $x_0 + x$, где $x$ принажлежит линейному подпространству $x \in V$, вектор $(x_0 + x) \in V_b$, значит $V_b - аффинное  подпространство.$

### Задача 9 (*)

Дан вектор $v\in \mathbb R^n$ (в виде одномерного `np.array`). Требуется написать программу, строящую матрицу, строками которой являются векторы $t v$ для $t = -1 + k/10,\ k = 0, \ldots, 20$ (то есть $t$ пробегает значения от $-1$ до $1$ с шагом $1/10$, всего $21$ значение). Получающаяся матрица должна иметь $21$ строку и $n$ столбцов. 

**Подсказка.** Для решения задачи вам нужно использовать матричное умножение, а для этого придётся делать из векторов матрицы (вектор-столбцы или вектор-строки). Это можно сделать, например, так:...

```python
def tv(array):
    n = 20
    t = np.diag([-1 + k/10. for k in range(n+1)])
    V = np.array([array for i in range(n+1)])
    return t @ V
```

```python
my_array = np.array([10, 20, 30, 40])
my_array
```

```python
c = tv(my_array)
c.shape
```

*Рангом* матрицы называется максимальное число её линейно независимых строк или максимальное число её линейно независимых столбцов (можно доказать, что оба числа обязательно совпадают).

Матрицы с малым рангом важны в задачах машинного обучения.

Пусть известно, что прямоугольная матрица $A$ имеет ранг $k$. Доказать, что можно выбрать $k$ строк и $k$ столбцов этой матрицы, таких, что по ним она однозначно восстанавливается. Найти явную формулу для такого восстановления. (Для решения этой задачи вам понадобится понятие обратной матрицы.)
